# Titanic Data that show passenger survived or not on the basis of data


```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df1=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df1.head()
```

```python
inputs=df1.drop(['Survived'],axis='columns')
inputs
```

```python
target=df1['Survived']
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
```

```python
inputs['sex_n']=le_sex.fit_transform(df['Sex'])
```

```python
inputs.head()
```

```python
input_n=inputs.drop(['Sex'],axis='columns')
```

```python
input_n.head()
```

```python
input_n.Age[:10]
```

```python
input_n.describe()
```

```python
input_n.Age=input_n.Age.fillna(input_n.Age.mean())
input_n.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_n,target,test_size=0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
